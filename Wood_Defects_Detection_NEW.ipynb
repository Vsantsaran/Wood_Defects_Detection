## YOLOv10b
### Training

```python
from ultralytics import YOLO
model = YOLO('best.pt')
model.train(data='data.yaml', epochs=40, batch=16, imgsz=640, workers=72, save=True, plots=True, cls=1.0)
```

```python
# Validation for train3
from ultralytics import YOLO
model = YOLO('runs/detect/train3/weights/best.pt')
model.val(data='data.yaml', batch=16, imgsz=640, workers=72, plots=True, save=True)
```

```python
# Prediction for train3
model.predict(batch=16, imgsz=640, workers=72, plots=True, source='Dataset1/test/images', save=True)
```
